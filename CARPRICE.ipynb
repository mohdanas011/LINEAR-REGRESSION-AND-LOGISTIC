```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Drop the 'name' column since it's not useful for numerical modeling
df = df.drop(columns=['name'])

# Encode categorical variables
label_encoders = {}
categorical_cols = ['fuel', 'seller_type', 'transmission', 'owner']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Store encoder for reference

# Define Features (X) and Targets (y)
X = df.drop(columns=['selling_price', 'owner'])  # Features
y_linear = df['selling_price']  # Target for Linear Regression
y_logistic = df['owner']  # Target for Logistic Regression

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data for Linear Regression
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(
    X_scaled, y_linear, test_size=0.2, random_state=42
)

# Split data for Logistic Regression
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_scaled, y_logistic, test_size=0.2, random_state=42
)

# Confirm preprocessing completion
X_train_lin.shape, X_train_log.shape
```

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score

# Train Linear Regression Model
lin_reg = LinearRegression()
lin_reg.fit(X_train_lin, y_train_lin)

# Predictions for Linear Regression
y_pred_lin = lin_reg.predict(X_test_lin)

# Evaluate Linear Regression
mae_lin = mean_absolute_error(y_test_lin, y_pred_lin)
mse_lin = mean_squared_error(y_test_lin, y_pred_lin)
rmse_lin = mse_lin ** 0.5

# Train Logistic Regression Model
log_reg = LogisticRegression(max_iter=500, random_state=42)
log_reg.fit(X_train_log, y_train_log)

# Predictions for Logistic Regression
y_pred_log = log_reg.predict(X_test_log)

# Evaluate Logistic Regression
accuracy_log = accuracy_score(y_test_log, y_pred_log)

# Output Results
(mae_lin, rmse_lin, accuracy_log)
```

```python
# Import necessary libraries
import numpy as np

# Define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate values for x-axis
x_values = np.linspace(-10, 10, 100)
y_values = sigmoid(x_values)

# Plot the sigmoid function
plt.figure(figsize=(8, 5))
plt.plot(x_values, y_values, label="Sigmoid Function", color="blue")
plt.axhline(y=0.5, color='red', linestyle="dashed", label="Decision Boundary (y=0.5)")
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
plt.title("S-shaped Sigmoid Function")
plt.legend()
plt.grid()
plt.show()
```
